## Modify element

In a list, you can modify an element by assigning its value directly to its index number.

```python
# make a copy
library_copy = library

print(library_copy)

# change library_copy index 2 into something else
library_copy[2] = "Book3333"

print(library_copy)
```

# Tuple

Very similar as list except once the element is in the collection, you <b>cannot change it</b>!

```python
# example of list syntax
library1 = ["book1", "book2", "book3", "book4", "book5", "book6", "book7"]
```

```python
# tuple syntax
library2 = ("book1", "book2", "book3", "book4", "book5", "book6", "book7")
```

```python
print(library1[4])

print(library2[4])
```

```python
print(type(library1))

print(type(library2))
```

```python
# we can also print the length of the tuple
print(len(library2))
```

<b>Question</b>: Can we modify the element inside a tuple?

Let's try it!

```python
print(type(library2))
print(library2)
```
