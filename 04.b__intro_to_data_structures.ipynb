## set

Very similar as list except you <b>cannot</b> have duplicate element(s)!

<b>Properties</b>:
- Unordered
- Changeable
- <b>Not allow duplicate elements</b>
- <b>Always separated by a comma between each elements</b>

```python
set1 = {"book1", "book2", "book3", "book4", "book5", "book6", "book7"}
```

```python
print(type(set1))
```

# dictionary

In the real world, we have a dictionary (see image below) where a single word holds/contains multiple information at once.



<b>Good news! We can do the same in Python!</b>



With dictionary data type, we always have key and value pair, and the key must be <b>unique</b>!

```python
python_class1 = {
    "student_1": "Nuoyu",
    "student_2": "Leo",
    "student_3": "JunRen",
    "student_4": "Whilliam",
    "student_5": "Daniel",
    "student_6": "Whilliam"
}

print(python_class1)
```

Alternately, we can use for loop to print it!

```python
for key, val in python_class1.items():
    print(f"{key}: {val}")
```

```python
print(python_class1["student_4"])
```

```python
print(python_class1["student_6"])
```

Let's try when we have duplicate keys.

```python
python_class2 = {
    "student_1": "Nuoyu",
    "student_1": "Leo",
    "student_3": "JunRen",
    "student_4": "Whilliam",
    "student_5": "Daniel",
    "student_6": "Whilliam"
}

print(python_class2)
```

```python
for key, val in python_class2.items():
    print(f"{key}: {val}")
```

Note, the key and value can be different data types (integer, float, boolean, and string) as we wish!

```python
dictionary1 = {
    0: "Hello World!",
    "Name": False,
    True: 1,
    0.22: 0.33
}
```

```python
print(dictionary1)
```

```python
for key, val in dictionary1.items():
    print(f"{key}: {val}")
```
