# Parameter

Sometimes, a function may accept a parameter(s).

Let's take an example of a function that calculates the area of a triangle:

$\Large A = \frac{1}{2}bh$

```python
def triangle_area(b, h):
    area = (1/2) * b * h
    return area

print(triangle_area(5, 3))
```

Therefore, $\Large A$ is the output of the function (LHS), while $\Large \frac{1}{2}bh$ is a function (RHS).

```python
# alternatively it can be rewritten as:
def triangle_area(b, h):
    return (1/2) * b * h

print(triangle_area(5, 3))
```
