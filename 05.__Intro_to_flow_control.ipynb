## Multiple Expressions

Sometimes, we may want to compare multiple expressions at the same time!

```python
# create a list that holds NYC weather in the next 3 days.
nyc_weather = [87, 90, 89]

# print day 1 weather
print(nyc_weather[0])

# print day 2 weather
print(nyc_weather[1])
```

```python
if nyc_weather[0] > 100 and nyc_weather[1] > 100:
    print("Super Hot!")
else:
    print("Not so bad.")
```

$\large \text{How can we change the operator(s) to make the above code print "Super Hot!" ?}$



```python
if nyc_weather[0] < 100 and nyc_weather[1] < 100:
    print("Super Hot!")
else:
    print("Not so bad.")
```

```python
if nyc_weather[0] > 100 and nyc_weather[1] > 100 and nyc_weather[2] > 100:
    print("Super Hot!")
else:
    print("Not so bad.")
```

<b>As we can see, we can have as many expressions as we wish! However, in the real world, we do not do that because it is a bad programming practice.</b>

## else if

Sometimes, we want to compare the multiple expressions but not necessary at the same time!

```python
# print day 1 weather
print(nyc_weather[0])

# print day 2 weather
print(nyc_weather[1])
```

```python
if nyc_weather[0] > 100:
    print("Day 1 is hot!")
elif nyc_weather[1] > 100:
    print("Day 2 is hot!")
else:
    print("Day 1 and day 2 are not hot!")
```

### <i>range()</i>

The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.

e.g., $[0, 1, 2, 3, 4, ..., n^{th}]$

```python
print(range(3))
```

```python
print(range(len(nyc_weather)))
```

### Visual Represenation 3

# While

Keep running a set of statements as long as a condition is true.

```python
number = 0

while number < 10:
    print("Hooray!")
    
    # increment by one
    number = number + 1
```

# <i>Methods: break, pass, continue</i> (optional)

Sometimes, we may want to something special to the loop:



```python
num = [9, 2, 4, 5, 2]
```

## break

```python
for item in num:
    print(item)
```

```python
for item in num:
    if item == 4:
        break
    
    print(item)
```

## pass

```python
for item in num:
    
```

```python
for item in num:
    pass
```

## continue

```python
for item in num:
    if item == 2:
        continue
    
    print(item)
```
