## Boolean

There will be a time where we want to compare more than two things at the time. Therefore, boolean operators will be helpful.

### <i>and</i>

Both options. Or, all of them.



```python
print("H" == "H" and "S" == "S")
```

```python
print("H" == "H" and "S" == "Z")
```

```python
print("H" == "Z" and "S" == "S")
```

```python
print("H" == "Z" and "S" == "Z")
```

### <i>or</i>

One or the two. Or, one of them.



```python
print("H" == "H" or "S" == "S")
```
